```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('healthcare_cleaned.csv')
df.head()
```

```python
# Set consistent style for all charts
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams["font.size"] = 12

print("Plot style set")
```

```python
# age distribution of patients
plt.figure(figsize=(10, 5))
sns.histplot(df["age"], bins=20, kde=True, color="steelblue")
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("age_distribution.png", dpi=150)
plt.show()

print("Avg Age:", df["age"].mean().round(2))
print("Min Age:", df["age"].min())
print("Max Age:", df["age"].max())
```

```python
#Age Category Distribution
age_counts = df["age_category"].value_counts()

plt.figure(figsize=(7, 5))
sns.barplot(x=age_counts.index, y=age_counts.values, palette="Blues_d")
plt.title("Patient Count by Age Category")
plt.xlabel("Age Category")
plt.ylabel("Number of Patients")
plt.tight_layout()
plt.savefig("age_category.png", dpi=150)
plt.show()

print(age_counts)
```

```python
# gender distribution
gender_counts = df["gender"].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index,
        autopct="%1.1f%%", colors=["#4C9BE8", "#F4845F"],
        startangle=90)
plt.title("Gender Distribution")
plt.tight_layout()
plt.savefig("gender_distribution.png", dpi=150)
plt.show()

print(gender_counts)
```

```python
#Medical Condition Distribution
condition_counts = df["medical_condition"].value_counts()

plt.figure(figsize=(10, 5))
sns.barplot(x=condition_counts.values, y=condition_counts.index, palette="coolwarm")
plt.title("Patients by Medical Condition")
plt.xlabel("Number of Patients")
plt.ylabel("Medical Condition")
plt.tight_layout()
plt.savefig("medical_conditions.png", dpi=150)
plt.show()

print(condition_counts)
```

```python
#Admission Type Distribution
admission_counts = df["admission_type"].value_counts()

plt.figure(figsize=(7, 5))
sns.barplot(x=admission_counts.index, y=admission_counts.values,
            palette="Set2")
plt.title("Admission Type Distribution")
plt.xlabel("Admission Type")
plt.ylabel("Number of Patients")
plt.tight_layout()
plt.savefig("admission_type.png", dpi=150)
plt.show()

print(admission_counts)
```

```python
# Billing Amount Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df["billing_amount"], bins=30, kde=True, color="green")
plt.title("Billing Amount Distribution")
plt.xlabel("Billing Amount")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("billing_distribution.png", dpi=150)
plt.show()

print("Avg Billing:", df["billing_amount"].mean().round(2))
print("Median Billing:", round(df["billing_amount"].median(), 2))
print("Max Billing:", df["billing_amount"].max())
```

```python
#Avg Billing by Medical Condition
avg_billing = df.groupby("medical_condition")["billing_amount"].mean().sort_values()

plt.figure(figsize=(10, 5))
sns.barplot(x=avg_billing.values, y=avg_billing.index, palette="magma")
plt.title("Average Billing Amount by Medical Condition")
plt.xlabel("Average Billing Amount")
plt.ylabel("Medical Condition")
plt.tight_layout()
plt.savefig("billing_by_condition.png", dpi=150)
plt.show()

print(avg_billing.round(2))
```

```python
#Length of Stay Analysis
plt.figure(figsize=(10, 5))
sns.histplot(df["length_of_stay"], bins=20, kde=True, color="purple")
plt.title("Length of Stay Distribution")
plt.xlabel("Days")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("length_of_stay.png", dpi=150)
plt.show()

print("Avg Stay:", df["length_of_stay"].mean().round(2), "days")
print("Max Stay:", df["length_of_stay"].max(), "days")
print("Min Stay:", df["length_of_stay"].min(), "days")
```

```python
# Order months correctly
month_order = ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November", "December"]

monthly = df["admission_month"].value_counts().reindex(month_order)

plt.figure(figsize=(12, 5))
sns.lineplot(x=monthly.index, y=monthly.values, marker="o", color="teal")
plt.title("Monthly Admission Trend")
plt.xlabel("Month")
plt.ylabel("Number of Admissions")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("monthly_trend.png", dpi=150)
plt.show()
```

```python
#Test Results Breakdown
test_counts = df["test_results"].value_counts()

plt.figure(figsize=(7, 5))
sns.barplot(x=test_counts.index, y=test_counts.values,
            palette=["green", "red", "orange"])
plt.title("Test Results Breakdown")
plt.xlabel("Test Result")
plt.ylabel("Number of Patients")
plt.tight_layout()
plt.savefig("test_results.png", dpi=150)
plt.show()

print(test_counts)
```

```python
# Insurance Provider Analysis
insurance_counts = df["insurance_provider"].value_counts()

plt.figure(figsize=(10, 5))
sns.barplot(x=insurance_counts.values, y=insurance_counts.index, palette="viridis")
plt.title("Patients by Insurance Provider")
plt.xlabel("Number of Patients")
plt.ylabel("Insurance Provider")
plt.tight_layout()
plt.savefig("insurance_provider.png", dpi=150)
plt.show()

print(insurance_counts)
```

```python
#Correlation Heatmap
plt.figure(figsize=(8, 6))
numeric_cols = df[["age", "billing_amount", "length_of_stay", "room_number"]]
sns.heatmap(numeric_cols.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig("correlation_heatmap.png", dpi=150)
plt.show()
```

```python
#Key Insights Summary
print("=" * 50)
print("KEY INSIGHTS FROM EDA")
print("=" * 50)

print(f"\n1. Most common condition : {df['medical_condition'].value_counts().idxmax()}")
print(f"2. Average patient age   : {df['age'].mean():.1f} years")
print(f"3. Most patients are     : {df['age_category'].value_counts().idxmax()}")
print(f"4. Average billing       : ₹{df['billing_amount'].mean():,.2f}")
print(f"5. Avg length of stay    : {df['length_of_stay'].mean():.1f} days")
print(f"6. Most common admission : {df['admission_type'].value_counts().idxmax()}")
print(f"7. Most used insurance   : {df['insurance_provider'].value_counts().idxmax()}")
print(f"8. Most common test result: {df['test_results'].value_counts().idxmax()}")

print("\n" + "=" * 50)
print("EDA Complete")
```
